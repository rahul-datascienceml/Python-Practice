**import**

- **import** is the keyword to use the packages.
- syntax is:
    - **import package_name**

- **ModuleNotFoundError:** Whenever the package is not available then we get this error.

**Directory**

- **dir** we notify a directory
- This tells us the details of the sub methods available in a package.
- syntax is:
    - **dir(package_name)**
    - ex. dir(pandas)



**Method Call**

- In order to call any method we need to call package first then method.
- syntax is:
      - **package_name.method_name** ex. random.randint
- If the method id not available in the package then we get **Attribute Error.**

**How to Use**

**help**

- In order to understand the method we need to first call the method and then we need to 'help' it.

- the syntax to understand the method:
    - **help(package_name.method_name)**
  - ex. help(random.randint) --> Return random integer in range [a, b], including both end points.
  - **Use of the method:** random.randint(1,6) --> It will give a random number between 1 and 6 including 1 and 6 as well.

```python
# improt package_name
# dir(package_name) it will give you the list of methods available in that package
# package_name.method_name
```

**How to install packages**

```python
#package name: cv2 : Image Read
             #Computer Vision
#package name: streamlit : This is used for creating apps
#package name: flask : flask is for API testing
```

- To install packages we need internet, if there is any issue with internet we will get **http** error.
- to install packages we need **terminal**
- In our case it is : **Anaconda Prompt**
- all the installation will need only one command **pip install package_name**
- **pip** full form is **preferred installation package**
- certain times installation package name and import package name might be different however in 90% cases it will be same.

- **Process - 1 to Install Package**
    - Open the **terminal** in Mac and check the directory you are currently in. In windows we need to open **Anaconda Prompt**
    - Google the package name and go to **PyPi** website to get the command to install the package.
    - comeback to terminal and paste the command, it will install the package.
    - after this we need to **restart the kernel.**

- **Process 2 to install Packages**
    - In Jupyter Notebook itself we can install the package.
    - To install : **!pip install package_name** (there is an exclamatory mark in before the command
